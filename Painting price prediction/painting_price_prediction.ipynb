```python
#so we can see that its not efficent to proceed with encoding as we have 3972 unique values
#and also if we enter the features in a different way it would not recognize
#so making them a list and appending to a list l and using set taking the individual elements and
#saving it to another list and making columns with that and filling 1 if its there in that row
```

```python
#We can see that some times they have wriiten it in capital letter and sometimes in small for same word so converting into lower
#splitting them with comma for making them come under a list
df.styles = df.styles.str.lower()
df.styles = df.styles.str.split(',')
df
```

```python
l = []
for row in df.styles:
    for i in row:
        l.append(i)
a = list(set(l))
len(a)
```

```python
#ignoring the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#creating rows with the style names and keeping value as 0
for indv_ele in a:
    df[indv_ele] = 0
df
```

```python
#looping and filling 1 if the style is present if not leaving it at 0 for the rows created
for element in a:
    for ind in range(len(df.styles)):
        if element in df.styles.iloc[ind]:
            df[element][ind] = 1
        else:
            continue
df
```

```python
#Storing the features and target variable 
x = df1.drop(['price'],axis = 1)
y = df1.price

#Taking 20% data for testing and remaining 80% for training
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```
