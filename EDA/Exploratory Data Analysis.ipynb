# Data cleaning

```python
for col in df:
    if df[col].isnull().sum()*100/len(df) >90:
        print(col)
```

```python
df['dismissal_kind'].unique()
```

```python
#Though dismissal kind have more that 90% of columns with nan, it is because they are not out
df['dismissal_kind'] = df['dismissal_kind'].fillna('not out')
```

```python
df['player_dismissed'].unique()
```

```python
#nan values are there because nobody was dismissied
df['player_dismissed'] = df['player_dismissed'].fillna('nobody')
```

```python
#removing fielder as lot of datas are misssing
df = df.drop(['fielder'],axis=1)
```

```python
#If there was no extra it is shown as nan, so changing that to no extra
df['extras_type'] = df['extras_type'].fillna('no extra')
```

```python
df
```

```python
df.isnull().sum()*100/len(df)
```

```python
"""We cannot impute the bowing team data with mean or median or mode as it is an important data and it will affect the analysis
and since it is only 0.01 percentage approx we remove those rows"""

cleaned_df = df.dropna()
cleaned_df
```

```python
cleaned_df.describe()
```

```python
#ball cannot be more than 6, so removing the ball greater than 6
cleaned_df = cleaned_df[cleaned_df['ball'] <= 6]
cleaned_df.describe()
```
